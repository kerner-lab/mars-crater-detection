```python
import numpy as np
import cv2
#BGR
color_dict = {'Other':(255,0,255),'Secondary':(0,0,255),'Buried':(255,0,0),'Layered':(0,255,0)}
def create_tile(MINLAT,MAXLAT,MINLON,MAXLON,new_df,files):
    print(MINLAT,MAXLAT,MINLON,MAXLON)
#     MINLAT = -30
#     MAXLAT = 0

#     MINLON = 0 
#     MAXLON = 30
    MINDIAM = 4
    MAXDIAM = 64
    f_df = new_df[(MINLAT<=new_df['LATITUDE_CIRCLE_IMAGE']) &(new_df['LATITUDE_CIRCLE_IMAGE']<=MAXLAT)
                  &(MINLON<=new_df['LONGITUDE_CIRCLE_IMAGE']) &(new_df['LONGITUDE_CIRCLE_IMAGE']<=MAXLON)
                 &(MINDIAM<=new_df['DIAM_CIRCLE_IMAGE']) &(new_df['DIAM_CIRCLE_IMAGE']<=MAXDIAM)]
    img_new = np.zeros((7680,7680,3), np.uint8)
    for i in range(len(f_df)):
        lat = f_df.iloc[i][1]
        # print(lat)
        lon =  f_df.iloc[i][2]
        diam =  f_df.iloc[i][5]
        # print(diam)
        if (MINLAT <= float(lat) and float(lat) <= MAXLAT and MINLON <= float(lon) and float(lon) <= MAXLON):
            # print('r')
            # print(lat,lon,diam)
            x_ctr = int(7680 * abs(lon - MINLON) / (MAXLON - MINLON))
            y_ctr = int(7680 * abs(MAXLAT - lat) / (MAXLAT - MINLAT))
            # y_ctr = int(abs(lat*256))
            r = int(diam*2.1608)
            brightness = color_dict[f_df.iloc[i]['STATUS']]
            # print(x_ctr,y_ctr,r,brightness)
            thick = -1
            #cv2.circle(img, center, radius, color, thickness=1, lineType=8, shift=0)
            img_new = cv2.circle(img_new,(x_ctr,y_ctr), r, brightness, thick) 
            # if(i==4):
            #     break
    # plt.imshow(img_new)
    # plt.show()
    cv2.imwrite('out/'+files.split('.')[0]+'.png',img_new)
```

```python
def loadIm(fname, tname,out_name,split, step=512, newpx = 512, px = 7680): 
    
    # im = plt.imread('Robbins_Dataset/out/'+fname)
    #px = 2048 #im.size #TODO: FIX THIS IT WILL BREAK EVERYTHING
    # print('max: ' + str(im.max()) + ', min: ' + str(im.min()) + ', mean: ' + str(im.mean()))
    # tim = 1*(plt.imread('Robbins_Dataset/out/'+tname)>0) #makes values of target binary
    im = cv2.imread('Robbins_Dataset/out/'+fname)
    print(tname)
    tim = cv2.imread(tname)
    counter = 0
    print(im.shape)
    print(tim.shape)
    count = 0
    for y in range(0,px,step): #no need to sub 512 b/c px are mult of 512
        for x in range(0,px,step):
            
            ipath = 'multiclass_data/'+split+'/images/'
            isExist = os.path.exists(ipath)
            if not isExist:
                os.makedirs(ipath)
            cv2.imwrite(ipath+out_name+str(count+step)+'.png',im[x:x+newpx,y:y+newpx].reshape((newpx,newpx,3)))
            mpath = 'multiclass_data/'+split+'/mask/'
            isExist = os.path.exists(mpath)
            if not isExist:
                os.makedirs(mpath)
            cv2.imwrite(mpath+out_name+str(count+step)+'.png',tim[x:x+newpx,y:y+newpx].reshape((newpx,newpx,3)))
            count += 1
            # break
```

```python
c = 0
for files in os.listdir('Robbins_Dataset/out/'):
    # if(files.split("_")[-1]=='segrng.png'):
    #     print(files)
    if(files.split("_")[-1]=='filled.png'):

        print(files)
        print(files[:-11]+'_2_32_km_segrng_8_edge.png')
        if(c<15):
            split = 'train'
        elif(c<23):
            split='val'
        else:
            split='test'
        tlist = (files.split("_")[4:8])
        for m in os.listdir('out/'):
            if(m.split("_")[3:7]==tlist):
                break
        print(m)
        loadIm(files,'out/'+m,files[:-11],split)
        c += 1
        # shutil.copy('Robbins_Dataset/out/'+files,'data/'+files)
```
